```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Read in the data from the potential xvg file
data = np.genfromtxt('potential.xvg', skip_header=25)

#Plot the data
plt.plot(data[:,0], data[:,1])
#Add a title and axis labels
plt.title('Lysozyme Potential Energy During Minimizatoin')
plt.xlabel('Time (ps)')
plt.ylabel('Potential Energy (kJ/mol)')

#Save the figure as a png in the directory '../Figures'
plt.savefig('../Figures/Lysozyme_Potential.png')


```

```python
#Read the data from the pressure xvg file, and plot a running average of the data
data = np.genfromtxt('pressure.xvg', skip_header=25)
#make an array of the running average of the last 10 data points
avg = []
for i in range(len(data[:,1])):
    if i < 10:
        avg.append(np.mean(data[:i,1]))
    else:
        avg.append(np.mean(data[i-10:i,1]))


#Plot the pressure and the running average of the pressure on the same plot
plt.plot(data[:,0], data[:,1], label='Pressure')
plt.plot(data[:,0], avg, label='Running Average - 10ps')
#Add a title and axis labels
plt.title('Lysozyme Pressure During NPT Equilibration')
plt.xlabel('Time (ps)')
plt.ylabel('Pressure (bar)')
#Add a legend
plt.legend()

#Save the figure as a png in the directory '../Figures'
plt.savefig('../Figures/Lysozyme_Pressure.png')




```

```python
#Do the same for density as you did for pressure
data = np.genfromtxt('density.xvg', skip_header=25)
avg = []
for i in range(len(data[:,1])):
    if i < 10:
        avg.append(np.mean(data[:i,1]))
    else:
        avg.append(np.mean(data[i-10:i,1]))

plt.plot(data[:,0], data[:,1], label='Density')
plt.plot(data[:,0], avg, label='Running Average - 10ps')
plt.title('Lysozyme Density During Equilibration')
plt.xlabel('Time (ps)')
plt.ylabel('Density (kg/m^3)')
plt.legend()

plt.savefig('../Figures/Lysozyme_Density.png')

```

```python
#Do the same for temperature as you did for pressure and density, but without running mean
data = np.genfromtxt('temperature.xvg', skip_header=25)

plt.plot(data[:,0], data[:,1], label='Temperature')
plt.title('Lysozyme Temperature During NVT Equilibration')
plt.xlabel('Time (ps)')
plt.ylabel('Temperature (K)')
plt.legend()

plt.savefig('../Figures/Lysozyme_Temperature.png')
```

```python
#Now we plot plot the RMSD of the protein backbone atoms along with the rmsd_xtal data on the same plot
data = np.genfromtxt('rmsd.xvg', skip_header=25)
data_xtal = np.genfromtxt('rmsd_xtal.xvg', skip_header=25)

plt.plot(data[:,0], data[:,1], label='RMSD')
plt.plot(data_xtal[:,0], data_xtal[:,1], label='RMSD Xtal')
plt.title('Lysozyme Backbone RMSD During Production')
plt.xlabel('Time (ps)')
plt.ylabel('RMSD (nm)')
plt.legend()

plt.savefig('../Figures/Lysozyme_RMSD.png')
```
