```python
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
#read in the data from Data/1hz3_T310.run.25000000.energy.xvg
data = np.genfromtxt(fname='../Data/1hz3_T310.run.25000000.energy.xvg', skip_header=29)
print(data)
```

```python
#Set the first column as the timestep
timestep = data[:,0]
#Convert time from ps to ns
timestep = timestep/1000
#Set the second column as the potential energy
potential_energy = data[:,1]
#Set the third column as the kinetic energy
kinetic_energy = data[:,2]
#Set the fourth column as the total energy
total_energy = data[:,3]
#Set the fifth column as the temperature
temperature = data[:,4]
#Set the sixth column as the pressure
pressure = data[:,5]
# Set the seventh column as the volume
volume = data[:,6]
```

```python
#Plot columns 2 through 7 as a function of the timestep separately
from turtle import color


fig, ax = plt.subplots(nrows=2, ncols=3, figsize=(10, 10))
ax[0,0].plot(timestep, potential_energy, color='red')
ax[0,0].set_xlabel('Timestep (ns)')
ax[0,0].set_ylabel('Potential Energy (kJ/mol)')
ax[0,0].set_title('Potential Energy vs Timestep')
ax[0,1].plot(timestep, kinetic_energy, color='blue')
ax[0,1].set_xlabel('Timestep (ns)')
ax[0,1].set_ylabel('Kinetic Energy (kJ/mol)')
ax[0,1].set_title('Kinetic Energy vs Timestep')
ax[0,2].plot(timestep, total_energy, color='green')
ax[0,2].set_xlabel('Timestep (ns)')
ax[0,2].set_ylabel('Total Energy (kJ/mol)')
ax[0,2].set_title('Total Energy vs Timestep')
ax[1,0].plot(timestep, temperature, color='orange')
ax[1,0].set_xlabel('Timestep (ns)')
ax[1,0].set_ylabel('Temperature (K)')
ax[1,0].set_title('Temperature vs Timestep')
ax[1,1].plot(timestep, pressure, color='purple')
ax[1,1].set_xlabel('Timestep (ns)')
ax[1,1].set_ylabel('Pressure (bar)')
ax[1,1].set_title('Pressure vs Timestep')
ax[1,2].plot(timestep, volume, color='black')
ax[1,2].set_xlabel('Timestep (ns)')
ax[1,2].set_ylabel('Volume (nm^3)')
ax[1,2].set_title('Volume vs Timestep')
fig.tight_layout()
```

```python
#Plot the potential energy and temperature as a function of the timestep on the same plot, with temperature on the left axis and potential energy on the right axis
fig, ax1 = plt.subplots(figsize=(10, 10))
ax1.plot(timestep, kinetic_energy, color='red')
ax1.set_xlabel('Timestep (ns)')
ax1.set_ylabel('Kinetic Energy (kJ/mol)', color='red')
ax1.tick_params(axis='y', labelcolor='red')
ax2 = ax1.twinx()
ax2.plot(timestep, potential_energy, color='orange')
ax2.set_ylabel('Potential Energy (kJ/mol)', color='orange')
ax2.tick_params(axis='y', labelcolor='orange')
fig.tight_layout()
#Add a title
ax1.set_title('Kinetic and Potential Energy vs Timestep')
```
