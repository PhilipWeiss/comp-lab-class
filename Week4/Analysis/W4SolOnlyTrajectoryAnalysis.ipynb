```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Read in the data from Oxygen_MSD.xvg from ../Data/OnlySol
OH2data = np.genfromtxt('../Data/OnlySol/WaterO_MSD.xvg', skip_header=23)
#Do the same for Sodium_MSD.xvg and Oxygen_MSD.xvg from ../Data/OnlySol
SodiumData = np.genfromtxt('../Data/OnlySol/Sodium_MSD.xvg', skip_header=23)
OxygenData = np.genfromtxt('../Data/OnlySol/Oxygen_MSD.xvg', skip_header=23)

#Plot the data vs timestep for all three on the same plot
plt.plot(OH2data[:,0], OH2data[:,1], label='OH2')
plt.plot(SodiumData[:,0], SodiumData[:,1], label='Sodium')
plt.plot(OxygenData[:,0], OxygenData[:,1], label='Oxygen')
plt.xlabel('Timestep (ps)')
plt.ylabel('MSD (nm^2)')
plt.legend()
#Add a title to the plot
plt.title('Mean Squared Displacement of Water, Sodium, and Oxygen')
plt.show()

#Save the plot as a png to the folder ../Figures
plt.savefig('../Figures/W4SolOnlyTrajectoryAnalysis.png')
```
